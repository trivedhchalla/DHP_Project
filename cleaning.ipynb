```python
# filling the NaN by using average ratio of the available data to the world average method
df = pd.read_csv('healthy_diet_cost.csv')
year_cols = [col for col in df.columns if col.isdigit()]
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
world_avg = df[year_cols].mean()
def fill_missing_with_ratio(row):
    available_years = row[year_cols].notna()
    if available_years.sum() == 0:
        return row
    ratios = row[year_cols][available_years] / world_avg[available_years]
    mean_ratio = ratios.mean()
    for year in year_cols:
        if pd.isna(row[year]):
            row[year] = world_avg[year] * mean_ratio
    return row
df_filled = df.apply(fill_missing_with_ratio, axis=1)
df_filled.to_csv('filled_healthy_diet_cost.csv', index=False)
```

```python
# filling of average wage data using average ratio of the available data to the world average method
df = pd.read_excel('avg_wage.xlsx')
year_cols = [col for col in df.columns if str(col).isdigit()]
df[year_cols] = df[year_cols].replace(0, np.nan)
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
world_avg = df[year_cols].mean()
def fill_missing_with_ratio(row):
    available_years = row[year_cols].notna()
    if available_years.sum() == 0:
        return row
    ratios = row[year_cols][available_years] / world_avg[available_years]
    mean_ratio = ratios.mean()
    for year in year_cols:
        if pd.isna(row[year]):
            row[year] = world_avg[year] * mean_ratio
    return row
df_filled = df.apply(fill_missing_with_ratio, axis=1)
df_filled.to_csv('filled_avg_wage.csv', index=False)
```

```python
# filling debt to gdp ratio data using average ratio of the available data to the world average method
df = pd.read_csv('debt_to_gdp_ratio.csv')
year_cols = [col for col in df.columns if str(col).isdigit()]
df[year_cols] = df[year_cols].replace(0, np.nan)
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
world_avg = df[year_cols].mean()
def fill_missing_with_ratio(row):
    available_years = row[year_cols].notna()
    if available_years.sum() == 0:
        return row
    ratios = row[year_cols][available_years] / world_avg[available_years]
    mean_ratio = ratios.mean()
    for year in year_cols:
        if pd.isna(row[year]):
            row[year] = world_avg[year] * mean_ratio
    return row
df_filled = df.apply(fill_missing_with_ratio, axis=1)
df_filled.to_csv('filled_debt_to_gdp_ratio.csv', index=False)
```

```python
# inflation rate adjustment same method average of ratio of availble data to world average 
df = pd.read_csv('Inflation_Rate.csv')
year_cols = [col for col in df.columns if str(col).isdigit()]
df[year_cols] = df[year_cols].replace(0, np.nan)
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
world_avg = df[year_cols].mean()
def fill_row(row):
    values = row[year_cols]
    available = values.notna()
    if available.sum() == 0:
        return row 
    ratio = values[available] / world_avg[available]
    mean_ratio = ratio.mean()
    row[year_cols] = values.fillna(world_avg * mean_ratio)
    return row

df_filled = df.apply(fill_row, axis=1)
df_filled.to_csv('filled_Inflation_Rate.csv', index=False)
```

```python
import pandas as pd

money_df = pd.read_csv('filled_avg_wage.csv')
exchange_df = pd.read_csv('exchangerate.csv')

money_df = money_df.loc[:, ['Country name'] + [str(year) for year in range(2006, 2023)]]
exchange_df = exchange_df.loc[:, ['Country Name'] + [str(year) for year in range(2006, 2023)]]

money_df.rename(columns={'Country name': 'Country'}, inplace=True)
exchange_df.rename(columns={'Country Name': 'Country'}, inplace=True)

merged = pd.merge(money_df, exchange_df, on='Country', suffixes=('_money', '_exchange'))

for year in range(2006, 2023):
    merged[str(year)] = merged[f'{year}_money'] / merged[f'{year}_exchange']

result = merged[['Country'] + [str(year) for year in range(2006, 2023)]]
result.to_csv('avg_wage.csv')
```

```python
# converting a raw json file into a procced good json file
def process_inflation_data(data):
    raw_entries = data[1] 

    country_data = {}

    for entry in raw_entries:
        country = entry['country']['value']
        year = int(entry['date'])
        value = entry['value']

        if value is None:
            continue

        if country not in country_data:
            country_data[country] = []

        country_data[country].append((year, value))
    for country in country_data:
        country_data[country].sort()
    return country_data
process_inflation_data(data)
```
